```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 8]

fortune = np.ones(100) * 100
people_id = range(100)

people_group1 = people_id[:10]
fortune_group1 = fortune[:10]
people_group2 = people_id[10:]
fortune_group2 = fortune[10:]
df1 = pd.DataFrame({"id": people_group1, "fortune": fortune_group1})
df2 = pd.DataFrame({"id": people_group2, "fortune": fortune_group2})
plt.bar(people_group1, fortune_group1, color="r")
plt.bar(people_group2, fortune_group2, color='b')
plt.show()

p = np.ones(100) * 1/(1.02*10+1*90)
for i in range(10):
    p[i] = 1/(1.02*10+1*90) * 1.02



days = 35*365
for i in range(days):
    fortune = fortune - np.ones(100)
    receivers = np.random.choice(people_id, 100, p=p)
    unique, counts = np.unique(receivers, return_counts=True)
    gains = zip(unique, counts)
    for (idx, gain) in gains:
        fortune[idx] += gain

print(fortune)

ind = np.argsort(fortune)[::-1]
rank = np.argsort(ind)

people_group1 = rank[:10]
fortune_group1 = fortune[:10]
people_group2 = rank[10:]
fortune_group2 = fortune[10:]
df1 = pd.DataFrame({"id": people_group1, "fortune": fortune_group1})
df2 = pd.DataFrame({"id": people_group2, "fortune": fortune_group2})
plt.bar(people_group1, fortune_group1, color="r")
plt.bar(people_group2, fortune_group2, color='b')
plt.show()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 8]

fortune = np.ones(100) * 100
people_id = range(100)

people_group1 = people_id[:10]
fortune_group1 = fortune[:10]
people_group2 = people_id[10:]
fortune_group2 = fortune[10:]
df1 = pd.DataFrame({"id": people_group1, "fortune": fortune_group1})
df2 = pd.DataFrame({"id": people_group2, "fortune": fortune_group2})
plt.bar(people_group1, fortune_group1, color="r")
plt.bar(people_group2, fortune_group2, color='b')
plt.show()

p = np.ones(100) * 1/(1.00*10+1*90)
spend = np.ones(100)

days = 35*365
for i in range(days):
    fortune = fortune - spend
    fortune[:10] = fortune[:10] + 0.01*(1+1e-4)**i
    receivers = np.random.choice(people_id, 100, p=p)
    unique, counts = np.unique(receivers, return_counts=True)
    gains = zip(unique, counts)
    for (idx, gain) in gains:
        fortune[idx] += gain


ind = np.argsort(fortune)[::-1]
rank = np.argsort(ind)

people_group1 = rank[:10]
fortune_group1 = fortune[:10]
people_group2 = rank[10:]
fortune_group2 = fortune[10:]
df1 = pd.DataFrame({"id": people_group1, "fortune": fortune_group1})
df2 = pd.DataFrame({"id": people_group2, "fortune": fortune_group2})
plt.bar(people_group1, fortune_group1, color="r")
plt.bar(people_group2, fortune_group2, color='b')
plt.show()
```
